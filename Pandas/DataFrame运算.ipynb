```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("./data/stock_day.csv")
data
```

## DataFrame运算

### 1. 算术运算

```python
# add()
# sub()
data['open'].add(1)
```

### 2. 逻辑运算

### 2.1 逻辑运算符号

```python
data[data["open"]>23].head()
```

```python
data[(data["open"] > 23) & (data["open"] < 24)].head()
```

### 2.2 逻辑运算函数

```python
# query(查询字符串)
data.query('open < 24 & open > 23').head()
```

```python
# isin(values)
data[data["open"].isin([23.53, 23.85])]  # 指定值进行一个判断，从而进行筛选操作
```

### 3.2 统计函数

```python
# median()：中位数
df = pd.DataFrame({'COL1' : [2,3,4,5,4,2],
                   'COL2' : [0,1,2,3,4,2]})

df.median()
```

```python
# 求出最大值的位置
data.idxmax(axis=0)
```

### 3.3 累计统计函数

cumsum : 计算前1/2/3/…/n个数的和

cummax : 计算前1/2/3/…/n个数的最大值

cummin : 计算前1/2/3/…/n个数的最小值

cumprod : 计算前1/2/3/…/n个数的积

```python
stock_rise = data['p_change']
stock_rise.cumsum()
```

### 4. 自定义运算

```python
# apply(func, axis=0)
# axis=0:默认是列，axis=1为行进行运算
# 定义一个对列，最大值-最小值的函数
data[['open', 'close']].apply(lambda x: x.max() - x.min(), axis=0)
```
