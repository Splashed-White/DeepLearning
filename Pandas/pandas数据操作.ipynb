```python
import pandas as pd
import numpy as np
```

## pandas数据操作

### 1. 文件读取

```python
data = pd.read_csv("./data/stock_day.csv")
```

```python
data
```

### 2. 索引操作

### 2.1 行列索引（先列后行）

```python
# data[列][行]
data['open']['2015-03-02']
```

### 2.2 loc / iloc

```python
# 使用loc:只能指定行列索引的名字
data.loc['2015-03-04':'2015-03-02','open']
```

```python
# 使用iloc可以通过索引的下标去获取
data.iloc[:3,:5]  # # 获取前3天数据,前5列的结果
```

### 2.3 ix组合索引

```python
# 使用ix进行下标和名称组合索引
data.ix[0:4,['open', 'close', 'high', 'low']]
```

```python
# 推荐使用loc和iloc来获取的方式
data.loc[data.index[0:4],['open', 'close', 'high', 'low']]
```

```python
data.iloc[0:4, data.columns.get_indexer(['open', 'close', 'high', 'low'])]
```

### 3. 赋值操作

```python
# 对DataFrame当中的close列进行重新赋值为1
data['close'] = 1
# data.close = 1
```

```python
data
```

### 4. 排序

### 4.1 DataFrame排序

```python
# df.sort_values(by=, ascending=）
# by:指定排序参考的建
# ascending(默认升序)  True = 升序，False = 降序
data.sort_values(by='open',ascending=True).head()
```

```python
# 使用df.sort_index给索引进行排序
data.sort_index()  # 日期重新排序
```

### 4.2 Series排序

```python
# series.sort_values(ascending=True)
data['p_change'].sort_values(ascending=True).head()
```

```python
# series.sort_index()
data['p_change'].sort_index().head()
```
